Data masking using lambda function:-

Applying the masking technique that helps us to make data replaces with characters instead of suppressing the records/columns for more number of count.
Here below we're applying the masking technique to the dt_customer column to make into hidden instead of deleting more number of rows or entire column

```python
dp['Dt_Customer'] = dp['Dt_Customer'].apply(lambda s: '##-##-####');
dp.head()
```

Data masking using faker package of python

Here we're masking the details with randomly generated names instead of originals for securing the names of both first and last for less loss of data

```python
from faker import Faker;
```

Assigning the facker constructor to a variable called fake_data.

```python
fake_data = Faker();
```

Using faker package we're faking both first and last names.

```python
dp['First Name'] = dp['First Name'].apply(lambda f: fake_data.name());
dp['First Name']
dp['Last Name'] = dp['Last Name'].apply(lambda f: fake_data.name());
dp['Last Name']
```

Same like fake names the emails and account numbers can also be generated 

```python
dp['email'] = dp['email'].apply(lambda v: fake_data.email())
dp['account_no'] = dp['account_no'].apply(lambda x: fake_data.bban())

dp['Mobile number'] = dp['Mobile number'].apply(lambda z: fake_data.msisdn()[3:])
dp.head()
```

After making the fake emails we've also masked them for making attacker to feel as secure data 

```python
dp['email'] = dp['email'].apply(lambda w: w[0] + "*****" + w[w.find('@'):]);
dp.head()
```

Data generalization

Here we're generalizing the data depending upon the outliers that exists according to the record/row wise and applying the top and bottom coding of binning technique.

Below is the top coding(right most) birth year outliers will be easily grab out by the attacker, so instead we're locking the age > 1990 to 1990 for saving subjects.

```python
#top coding on Year_Birth
dp.loc[dp['Year_Birth'] > 1990, 'Year_Birth'] = 1990;
dp['Year_Birth'].hist(bins=30);
```

Now let's see bottom coding(left most of above histogram) which is reverse of the top coding

```python
#bottom coding on above hist
dp.loc[dp['Year_Birth'] < 1950, 'Year_Birth'] = 1950;
dp['Year_Birth'].hist(bins=30);
```

Now we can see no outliers in above newly generated Year_Birth attribute

```python
key = Fernet.generate_key()
 
# string the key in a file
with open('filekey.key', 'wb') as filekey:
   filekey.write(key)
```

```python
with open('filekey.key', 'rb') as filekey:
    key = filekey.read()
 
# using the generated key
fernet = Fernet(key)
 
# opening the original file to encrypt
with open('out_test.csv', 'rb') as file:
    original = file.read()
     
# encrypting the file
encrypted = fernet.encrypt(original)
 
# opening the file in write mode and
# writing the encrypted data
with open('out_test.csv', 'wb') as encrypted_file:
    encrypted_file.write(encrypted)
```
